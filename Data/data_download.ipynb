```python
import urllib.request
import random
import os
import time
```

```python
# Generate a pool of 100 valid arXiv IDs from January 2014 (format: 1401.NNNN)
# We use this to guarantee the papers actually exist on arXiv's servers.
docbank_arxiv_pool = [f"1401.{str(i).zfill(4)}" for i in range(1, 101)]

# Select 30 random papers
selected_ids = random.sample(docbank_arxiv_pool, 30)

successful_downloads = 0

print(f"Attempting to download 20 random papers...")
print("=" * 80)

# Download and display the PDFs
for arxiv_id in selected_ids:
    pdf_url = f"https://arxiv.org/pdf/{arxiv_id}.pdf"
    file_name = f"docbank_{arxiv_id}.pdf"
    
    print(f"Downloading {file_name} from {pdf_url}...")
    
    try:
        # Adding a custom user-agent to prevent being blocked by arXiv for automated scraping
        req = urllib.request.Request(pdf_url, headers={'User-Agent': 'Mozilla/5.0'})
        
        with urllib.request.urlopen(req) as response, open(file_name, 'wb') as out_file:
            out_file.write(response.read())
            
        print(f"✅ Successfully downloaded {file_name}.")
        successful_downloads += 1
        
    except urllib.error.HTTPError as e:
        print(f"❌ Failed to download {arxiv_id} (HTTP Error: {e.code})")
    except Exception as e:
        print(f"❌ Failed to download {arxiv_id}: {e}")
        
    print("-" * 80)
    
    # Be nice to arXiv's servers by adding a tiny delay between downloads
    time.sleep(1)

print(f"\nFinished! Successfully downloaded and displayed {successful_downloads} out of 30 papers.")
```
